```python
# start with pmid data
with open(path+R'processed_data\word_id_dict_+2_tfidf.json') as json_string:
    word_id_dict = json.load(json_string)
```

```python
word_list = list(df_communities['word'].dropna())

pubs = []
communities = []
words = []

for word in word_list:
    community = list(df_communities['community'][df_communities['word'] == word])[0]
    for pub_id in word_id_dict[word]:
        pubs.append(pub_id)
        communities.append(community)
        words.append(word)
    
```

```python
df_communities['community'][df_communities['word'] == word]
```

```python
len(set(pubs))
```

```python
df_sem = pd.DataFrame(zip(pubs, communities, words)).reset_index().rename(columns = {1:'community', 0: 'id_select', 2:'word'})

grouped = df_sem.groupby(['id_select','community']).size().reset_index(name='count')
most_frequent = grouped.loc[grouped.groupby('id_select')['count'].idxmax()]
```

```python
words_sel_dict = {}
for pub in set(df_sem['id_select']): 
    words_sel = list(df_sem['word'][df_sem['id_select'] == pub])
    words_sel_str = ''
    for word in words_sel: 
        if words_sel_str == '': 
            words_sel_str = word
        else: 
            words_sel_str = words_sel_str+', ' +word
    words_sel_dict[pub] = words_sel_str
    
```

```python
most_frequent['words_sel'] = most_frequent['id_select'].map(words_sel_dict)
```

```python

df_grtype_scholarly_filt = df_grtype_scholarly[['pub_id_sel','GROUP_dsi','GROUP_lit']]
df_grtype_scholarly_semantics = pd.merge(df_grtype_scholarly, most_frequent[['pub_id_sel','sem_community']], on = 'pub_id_sel', how = 'inner')
```

```python

semantic_dict = {
    10: 'Public health',
    4: 'Evolutionary biology',
    1: 'Environmental microbiology',
    2: 'Molecular biology',
    14: 'Species identification',
    6: 'Agriculture and diseases',
    -2: 'No semantic leaning'
}

df_grtype_scholarly_semantics['semantic_name'] = df_grtype_scholarly_semantics['sem_community'].map(semantic_dict)

# semantics
# Get unique types of 'semantics' after dropping NaNs
semantics = sorted(set(df_grtype_scholarly_semantics['semantic_name'].dropna()))

# Number of subplots
n = len(semantics)

# Calculate rows and columns for the grid
n_cols = 2
n_rows = (n + 1) // n_cols  # Ensures correct number of rows for odd n

# Create subplots with 2 rows and multiple columns
fig, axes = plt.subplots(n_rows, n_cols, figsize=(10, 10), sharey=False)

# Flatten axes array for easier iteration
axes = axes.flatten()

# Set the theme for the plot
sns.set_theme(style="darkgrid")

# Define the palette for the different groups
palette = {"G77": "tab:purple", "OECD": "tab:blue", "BRICS": "tab:red"}

for i, semantic in enumerate(semantics):
    vis_df_subgroup = df_grtype_scholarly_semantics[df_grtype_scholarly_semantics['semantic_name'] == semantic]
    vis_df_subgroup_count1pg = vis_df_subgroup[['pub_id_sel','GROUP_dsi','GROUP_lit']].drop_duplicates().rename(columns = {'GROUP_dsi':'GROUP_dsi'}).reset_index(drop=True).sort_values(['GROUP_lit'], ascending = False)

    sns.countplot(
        data=vis_df_subgroup_count1pg,
        x="GROUP_dsi", 
        hue="GROUP_lit",
        order = ['OECD','BRICS','G77'],
        palette=palette, 
        ax=axes[i]
    )

    # Set labels and title
    axes[i].set_title(semantic, fontsize=14)
    axes[i].set_xlabel('GROUP_dsi (providers)', fontsize=12)
    axes[i].set_ylabel('Count', fontsize=12)
    axes[i].tick_params(axis='x', labelsize=10)
    axes[i].tick_params(axis='y', labelsize=10)

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()

# Save the combined figure
plt.savefig(path + R'plots\coo_cor_combined_semantic.svg', format="svg")
plt.savefig(path + R'plots\coo_cor_combined_semantic.jpg', format="jpg")

plt.show()
```
