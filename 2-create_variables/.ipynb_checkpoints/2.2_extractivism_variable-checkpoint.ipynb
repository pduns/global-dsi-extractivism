```python
#get group of cor
df_country2grp = pd.read_csv(path+R'raw_data\2024-08-25_country2grp.csv.gz',keep_default_na=False, na_values=[''])
df_country = pd.read_csv(path+R'raw_data\2024-08-25_country.csv.gz',keep_default_na=False, na_values=[''])

#rduce to main groups to simplify
main_groups = ['G77', 'OECD', 'BRICS']
df_country2grp_ed = df_country2grp.loc[df_country2grp['GRP'].isin(main_groups)]

#create dictionaries: (country name -->) iso --> group
country_iso = dict(zip(list(df_country['NAME']),list(df_country['ISO_STATE'])))

iso_group_dict = dict(zip(list(df_country2grp_ed['ISO']),list(df_country2grp_ed['GRP'])))
iso_group_dict['IN'] = 'BRICS'
iso_group_dict['ZA'] = 'BRICS'
iso_group_dict['BR'] = 'BRICS'
iso_group_dict['CL'] = 'G77'
iso_group_dict['CO'] = 'G77'
iso_group_dict['CR'] = 'G77'




df_scholarly_country['GROUP_lit'] = df_scholarly_country['affiliation_country'].map(iso_group_dict)
df_scholarly_country['ISO_ori'] = df_scholarly_country['COUNTRY'].map(country_iso)
df_scholarly_country['GROUP_ori'] = df_scholarly_country['ISO_ori'].map(iso_group_dict)
```

```python
#somehow drop nans before running the next step
df_scholarly_country_nn_np = df_scholarly_country[['id_select','GROUP_ori','GROUP_lit','first_author','last_author']].dropna().drop_duplicates()
```

```python
len(set(df_scholarly_country['id_select']))
```

```python
len(set(df_scholarly_country_nn_np['id_select']))
```

```python
# Initialize lists to hold the results
extracts_oecd_1_test = [] #resource from G77, 1 of last or first author from OECD
extracts_brics_1_test = [] #resource from G77, 1 of last or first author from BRICS
extracts_oecd_2_test = [] #resource from G77, all authors from OECD
extracts_brics_2_test = [] #resource from G77, all authors from BRICS
coorps_oecd_test = [] #cooperation between G77 and OECD if extracts_1 is not happening
coorps_brics_test = [] #cooperation between G77 and BRICs if extracts_3 is not happening 


# Group the DataFrame by 'id_select'
grouped_df = df_scholarly_country_nn_np.groupby('id_select')

for pub, group in grouped_df:
    origin_groups = set(group['GROUP_ori'].dropna())
    first_author_group = set(group['GROUP_lit'][(group['first_author'] == 1)])
    last_author_group = set(group['GROUP_lit'][(group['last_author'] == 1)])
    first_last_author_groups = first_author_group.union(last_author_group)
    author_groups = set(group['GROUP_lit'])

    if first_last_author_groups == set() and len(author_groups) == 1:
        first_last_author_groups = author_groups
        
        
        
    #if first or last author known, label
    if len(first_last_author_groups) >= 1: 
        extract_oecd_1 = int('G77' in origin_groups and 'OECD' in first_last_author_groups)
        extract_brics_1 = int('G77' in origin_groups and 'BRICS' in first_last_author_groups)
    #else label as unknown
    else: 
        extract_oecd_1 = np.nan
        extract_brics_1 = np.nan
        print(pub)
        
    #label everything regardless of whether first and last are known
    extract_oecd_2 = int('G77' in origin_groups and len(author_groups) == 1 and 'OECD' in author_groups)
    extract_brics_2 = int('G77' in origin_groups and len(author_groups) == 1 and 'BRICS' in author_groups)
    
    if not pd.isna(extract_oecd_1):
        coorp_oecd = int('G77' in author_groups and 'OECD' in author_groups and extract_oecd_1 == 0 and extract_oecd_2 == 0)
        coorp_brics = int('G77' in author_groups and 'BRICS' in author_groups and extract_brics_1 == 0 and extract_brics_2 == 0)
    else:
        coorp_oecd   = np.nan
        coorp_brics  = np.nan
        
    
    #print(pub)
    #print('origin_groups: '+str(origin_groups))
    #print('first_last_author_groups: '+str(first_last_author_groups))
    #print('author_groups: '+str(author_groups))
    #print('extract_oecd_1: '+str(extract_oecd_1))
    #print('extract_brics_1: '+str(extract_brics_1))
    #print('extract_oecd_2: '+str(extract_oecd_2))
    #print('extract_brics_2: '+str(extract_brics_2))
    #print('coorp_oecd: '+str(coorp_oecd))
    #print('coorp_brics: '+str(coorp_brics))
    
    extracts_oecd_1_test.append(extract_oecd_1)
    extracts_brics_1_test.append(extract_brics_1)
    extracts_oecd_2_test.append(extract_oecd_2)
    extracts_brics_2_test.append(extract_brics_2)
    coorps_oecd_test.append(coorp_oecd)
    coorps_brics_test.append(coorp_brics)


# Create DataFrame directly from the group key (id_select) and the calculated lists
df_extract_test = pd.DataFrame({
    'id_select': grouped_df['id_select'].first().index,
    'extractivism_oecd_1': extracts_oecd_1_test,
    'extractivism_brics_1': extracts_brics_1_test,
    'extractivism_oecd_2': extracts_oecd_2_test,
    'extractivism_brics_2': extracts_brics_2_test,
    'cooperation_oecd_1': coorps_oecd_test,
    'cooperation_brics_1': coorps_brics_test
})
```
